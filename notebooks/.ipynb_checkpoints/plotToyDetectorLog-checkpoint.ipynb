```python
from pylab import *
import pandas as pd
from sklearn import datasets
%matplotlib inline
```

```python
#machine = 'ryzen'
machine = "patatrack02"
with open("../log/{}_toyDetector_100layer.log".format(machine)) as f:
    lines = f.readlines()
runtimes = []
for line in lines:
    if line == 'Start to run CLUE algorithm\n':
        runtime = []
    if line == 'Finished running CLUE algorithm\n':
        runtimes.append(runtime)
    if line[0:13]=="Elapsed time:":
        runtime.append(float(line[13:-3]))
cpu100 = np.array(runtimes[ 0:10]) [:,3:]
gpu100 = np.array(runtimes[10:20]) [:,3:]
cuplaTBB100 = np.array(runtimes[ 20:30]) [:,3:] 
cuplaSerial100 = np.array(runtimes[ 30:40]) [:,3:] 
cuplaCuda100 = np.array(runtimes[ 40:50]) [:,3:] 

with open("../log/{}_toyDetector_1layer.log".format(machine)) as f:
    lines = f.readlines()
runtimes = []
for line in lines:
    if line == 'Start to run CLUE algorithm\n':
        runtime = []
    if line == 'Finished running CLUE algorithm\n':
        runtimes.append(runtime)
    if line[0:13]=="Elapsed time:":
        runtime.append(float(line[13:-3]))
cpu1 = np.array(runtimes[ 0:10])[:,3:] 
gpu1 = np.array(runtimes[10:20])[:,3:]
cuplaTBB1 = np.array(runtimes[ 20:30]) [:,3:] 
cuplaSerial1 = np.array(runtimes[ 30:40]) [:,3:] 
cuplaCuda1 = np.array(runtimes[ 40:50]) [:,3:] 
```

```python
plt.figure(figsize=(12,4),facecolor='w')
n = np.arange(1,11,1)

plt.subplot(121)
plt.errorbar(n, cpu1.mean(axis=1),yerr=cpu1.std(axis=1), marker='o',linestyle='--',label='CPU')
plt.errorbar(n, gpu1.mean(axis=1),yerr=gpu1.std(axis=1), marker='o',linestyle='--',label=r'GPU')
plt.errorbar(n, cuplaCuda1.mean(axis=1),yerr=cuplaCuda1.std(axis=1), marker='o',linestyle='--',label='CUPLA CUDA')
plt.errorbar(n, cuplaTBB1.mean(axis=1),yerr=cuplaTBB1.std(axis=1), marker='o',linestyle='--',label='CUPLA TBB')
plt.errorbar(n, cuplaSerial1.mean(axis=1),yerr=cuplaSerial1.std(axis=1), marker='o',linestyle='--',label='CUPLA Serial')
plt.grid(linestyle='--',alpha=0.5)
plt.legend()
plt.xlabel(r"Number of Points per Layer [$10^3$]")
plt.ylabel("Total Execution Time [ms]")
plt.grid(linestyle='--',alpha=0.5)
plt.legend()
plt.xlabel(r"Number of Points per Layer [$10^3$]")
plt.ylabel("Total Execution Time [ms]")
plt.title("1 layer")

plt.subplot(122)
plt.errorbar(n, cpu100.mean(axis=1),yerr=cpu100.std(axis=1), marker='o',linestyle='--',label='CPU')
plt.errorbar(n, gpu100.mean(axis=1),yerr=gpu100.std(axis=1), marker='o',linestyle='--',label=r'GPU')
plt.errorbar(n, cuplaCuda100.mean(axis=1),yerr=cuplaCuda100.std(axis=1), marker='o',linestyle='--',label='CUPLA CUDA')
plt.errorbar(n, cuplaTBB100.mean(axis=1),yerr=cuplaTBB100.std(axis=1), marker='o',linestyle='--',label='CUPLA TBB')
plt.errorbar(n, cuplaSerial100.mean(axis=1),yerr=cuplaSerial100.std(axis=1), marker='o',linestyle='--',label='CUPLA Serial')
plt.grid(linestyle='--',alpha=0.5)
plt.legend()
plt.xlabel(r"Number of Points per Layer [$10^3$]")
plt.ylabel("Total Execution Time [ms]")
plt.title("100 layers")

plt.savefig("../plots/{}_executionTime.png".format(machine),dpi=300)
```

```python
with open("../log/{}_toyDetector_100layer_nvprof.log".format(machine)) as f:
    lines = f.readlines()
    
HToD,DtoH,k1,k2,k3,k4,k5,memset = [],[],[],[],[],[],[],[]
for line in lines:  
    wordsInLine = line.split("  ")
    
    if wordsInLine[-1] == "[CUDA memcpy HtoD]\n":
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        HToD.append(time)
        
    if wordsInLine[-1] == "[CUDA memcpy DtoH]\n":
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        DtoH.append(time)
        
    if wordsInLine[-1] == "cudaMemset\n":
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        memset.append(time)
                
        
    name = "kernel_compute_histogram"
    if wordsInLine[-1][:len(name)] == name:
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        k1.append(time)
            
    name = "kernel_compute_density"
    if wordsInLine[-1][:len(name)] == name:
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        k2.append(time)
    
    name = "kernel_compute_distanceToHigher"
    if wordsInLine[-1][:len(name)] == name:
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        k3.append(time)
        
    name = "kernel_find_clusters"
    if wordsInLine[-1][:len(name)] == name:
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        k4.append(time)
        
    name = "kernel_assign_clusters"
    if wordsInLine[-1][:len(name)] == name:
        word = wordsInLine[-9]
        time = float(word[:-2])
        if word[-2:] == "us":
            time /= 1000
        k5.append(time)
        
prof100 = np.array([HToD,DtoH,memset,k1,k2,k3,k4,k5])/10
```

```python


fig, ax1 = plt.subplots(figsize=(8,5),facecolor='w')
ax2 = ax1.twinx()

n = np.arange(1,11,1)




ax1.errorbar(n, cuplaCuda100.mean(axis=1),yerr=cuplaCuda100.std(axis=1), marker='*',linestyle='--',color='C3',label=r'CUPLA CUDA')
ax1.errorbar(n, gpu100.mean(axis=1),yerr=gpu100.std(axis=1), marker='o',linestyle='--',color='C1',label=r'CUDA')



prof = prof100
but = np.cumsum(prof,axis=0)

labels = [
          "-- cuda memcpy HToD",
          "-- cuda memcpy DToH",
          "-- cuda memset",
          "-- build tiles",
          "-- calculate density",
          "-- calculate separation",
          "-- promote seeds",
          "-- assign clusters",
          "-- other"
         ]

colors = ["C1", "C2", "C7","C3","C0","C5", "C6","C4", "c" ]


ax1.bar(n, prof[0],label= labels[0],width=0.4, facecolor=colors[0])
ax1.bar(n, prof[1], bottom=but[0],label= labels[1],width=0.4, facecolor=colors[1] )
for i in range(2,8):
    ax1.bar(n, prof[i], bottom=but[i-1],label= labels[i],width=0.8, facecolor=colors[i] )
other = gpu100.mean(axis=1) - but[7]
ax1.bar(n, other, bottom=but[7],width=0.4,label= labels[8], facecolor=colors[8] )

ax1.set_ylim(0,15)
ax1.set_xlim(0,11)
ax1.grid(linestyle='--',alpha=0.5)
ax1.legend(fontsize=9)

ax1.set_xlabel(r"Number of Points per Layer [$10^3$]",fontsize=12)
ax1.set_ylabel("GPU Total Execution Time [ms]",fontsize=12,color="C1")
ax1.tick_params(axis='y', labelcolor="C1")
ax1.set_title("100 layers. Tested on AMD Ryzen 2700X [16T] + NVIDIA GTX 1080Ti",fontsize=12)
ax1.set_title("100 layers. Tested on Intel Xeon Silver 4114 [40T]+ NVIDIA Tesla V100 ",fontsize=12)

ax2.errorbar(n, cpu100.mean(axis=1),yerr=cpu100.std(axis=1), marker='o',linestyle='--',color='k',label=r'CPU Serial [1T]')
ax2.errorbar(n, cuplaSerial100.mean(axis=1),yerr=cuplaSerial100.std(axis=1), marker='^',linestyle='--',color='0.5',label=r'CUPLA CPU Serial [1T]')
ax2.errorbar(n, cuplaTBB100.mean(axis=1),yerr=cuplaTBB100.std(axis=1), marker='s',linestyle='--',color='0.3',label=r'CUPLA TBB [40T]')
ax2.legend(fontsize=9,loc="upper center")
ax2.set_ylim(0,800)
ax2.set_ylabel("CPU Total Execution Time [ms]",fontsize=12)



plt.savefig("../plots/{}_executionTime2.png".format(machine),dpi=300)
```
