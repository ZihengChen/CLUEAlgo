```python
from pylab import *
import pandas as pd
from sklearn import datasets
%matplotlib inline
```

# 1. Load Points

```python
n,m,nNoise = 25,10,5
k = int((n-nNoise)/m)

# load data
hits = np.genfromtxt("../data/input/test_25.csv",delimiter=",")
```

# 2. Run CLUE Clustering

```python
#####################################
# calculate local density
#####################################
rhos = []
for i in range(n):
    dist = np.sum(((hits[:,0:2]-hits[i,0:2])**2),axis=1)**0.5
    rhos.append(np.sum(dist<1))
rhos = array(rhos)



#####################################
# nearest higher
#####################################
nhs = []
deltas = []
for i in range(n):
    delta = 100
    nh = -1
    for j in range(n):
        if (rhos[j] > rhos[i]) | (( rhos[j]==rhos[i]) & (j>i)):

            dist = np.sum((hits[j,0:2]-hits[i,0:2])**2)**0.5
            if dist<=2:
                if dist < delta:
                    delta = dist
                    nh = j
    deltas.append(delta)
    nhs.append(nh)

nhs = np.array(nhs)
deltas = np.array(deltas)


#####################################
# Seed and outliers
#####################################
isSeeds = np.logical_and(deltas>2 , rhos >3)
isNoises = np.logical_and(deltas>2 , rhos <=3)
# isHits = np.logical_not(np.logical_or(isNoises,isSeeds))
isHits = np.logical_not((isNoises))

seeds = np.arange(isSeeds.size)[isSeeds]

#####################################
# Assign
#####################################

clusIds = -1*np.ones_like(nhs)


for i in range(seeds.size):
    clusIds[seeds[i]] = i

stack = list(seeds)
while len(stack) >0:
    i = stack.pop()
    iclusId = clusIds[i]
    ifollowers = np.arange(n)[nhs==i]
    for j in ifollowers:
        clusIds[j] = iclusId
        stack.append(j)
```

# 3. Plot Figure 2

```python
fig, axes = plt.subplots(1,4,figsize=(16,4),facecolor='w')
fig.subplots_adjust(wspace=0.08)

#####################################
ax = axes[0]
ax.set_aspect('equal')
ax.scatter(hits[:,0],hits[:,1],marker='o',s=50*rhos,c=rhos, linewidth=0.5, edgecolor="k", alpha=1,cmap='rainbow',vmin=0, vmax=6)
ax.grid(True,color='grey',linestyle="--",alpha=0.4)
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.tick_params(axis='y',labelsize=0,labelcolor="w")
ax.tick_params(axis='x',labelsize=0,labelcolor="w")
ax.text(-5,4.5,r"$a$",fontsize=18)


#####################################
ax = axes[1]
ax.set_aspect('equal')
ax.scatter(hits[:,0],hits[:,1],marker='o',s=50*rhos,c=rhos,  linewidth=0.5, edgecolor="k",cmap='rainbow', alpha=1, vmin=0, vmax=6)
for i in range(n):
    nh = int(nhs[i])
    if nh>=0:
        xx = array([ hits[i,0], hits[nh,0] ])
        yy = array([ hits[i,1], hits[nh,1] ])
        ax.plot(xx, yy,'k',linewidth=1)
        
        xarrow = 0.2*(xx[1]-xx[0])+xx[0]
        yarrow = 0.2*(yy[1]-yy[0])+yy[0]
        
        if (xx[1]-xx[0]) == 0:
            ax.arrow( xarrow, yarrow, 0 , -0.1*sign(yy[1]-yy[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
        if (xx[1]-xx[0]) > 0:
            ax.arrow( xarrow, yarrow, -0.1 , -0.1*(yy[1]-yy[0])/(xx[1]-xx[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
        else:
            ax.arrow( xarrow, yarrow, 0.1 , 0.1*(yy[1]-yy[0])/(xx[1]-xx[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
ax.grid(True,color='grey',linestyle="--",alpha=0.4)
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.tick_params(axis='y',labelsize=0,labelcolor="w")
ax.tick_params(axis='x',labelsize=0,labelcolor="w")
ax.text(-5,4.5,r"$b$",fontsize=18)


#####################################
ax = axes[2]
ax.set_aspect('equal')

ax.scatter(hits[isHits,0],hits[isHits,1],marker='o',s=60*rhos[isHits], c=rhos[isHits], linewidth=0.5,cmap='rainbow', edgecolor="k", alpha=1, vmin=0, vmax=6)

for i in range(n):
    nh = int(nhs[i])
    if (nh>=0 and isHits[i]) and not isSeeds[i]:
        xx = [ hits[i,0], hits[nh,0] ]
        yy = [ hits[i,1], hits[nh,1] ]
        ax.plot(xx,yy,'k',linewidth=1)
        
        xarrow = 0.2*(xx[1]-xx[0])+xx[0]
        yarrow = 0.2*(yy[1]-yy[0])+yy[0]
        if (xx[1]-xx[0]) == 0:
            ax.arrow( xarrow, yarrow, 0 , -0.1*sign(yy[1]-yy[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
        if (xx[1]-xx[0]) > 0:
            ax.arrow( xarrow, yarrow, -0.1 , -0.1*(yy[1]-yy[0])/(xx[1]-xx[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
        else:
            ax.arrow( xarrow, yarrow, 0.1 , 0.1*(yy[1]-yy[0])/(xx[1]-xx[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
             
ax.scatter(hits[isNoises,0],hits[isNoises,1],marker='s',s=10,c='grey', alpha=0.8)
ax.scatter(hits[8,0],hits[8,1],marker='*',s=800,c='g', linewidth=0.5,edgecolor="k",alpha=0.8)
ax.scatter(hits[17,0],hits[17,1],marker='*',s=800,c='r', linewidth=0.5,edgecolor="k",alpha=0.8)

ax.grid(True,color='grey',linestyle="--",alpha=0.4)
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.tick_params(axis='y',labelsize=0,labelcolor="w")
ax.tick_params(axis='x',labelsize=0,labelcolor="w")
ax.text(-5,4.5,r"$c$",fontsize=18)


#####################################
ax = axes[3]
ax.set_aspect('equal')

ax.scatter(hits[clusIds==0,0],hits[clusIds==0,1] ,marker='o',s=100, c="g", linewidth=0.5,edgecolor="k",alpha=1)
ax.scatter(hits[clusIds==1,0],hits[clusIds==1,1] ,marker='o',s=100, c="r", linewidth=0.5,edgecolor="k",alpha=1)
for i in range(n):
    nh = int(nhs[i])
    if (nh>=0 and isHits[i]) and not isSeeds[i]:
        xx = [ hits[i,0], hits[nh,0] ]
        yy = [ hits[i,1], hits[nh,1] ]
        ax.plot(xx,yy,'k',linewidth=1)
        
        xarrow = 0.2*(xx[1]-xx[0])+xx[0]
        yarrow = 0.2*(yy[1]-yy[0])+yy[0]
        if (xx[1]-xx[0]) == 0:
            ax.arrow( xarrow, yarrow, 0 , -0.1*sign(yy[1]-yy[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
        if (xx[1]-xx[0]) > 0:
            ax.arrow( xarrow, yarrow, -0.1 , -0.1*(yy[1]-yy[0])/(xx[1]-xx[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
        else:
            ax.arrow( xarrow, yarrow, 0.1 , 0.1*(yy[1]-yy[0])/(xx[1]-xx[0]), shape='full', lw=1, length_includes_head=True, head_width=0.2,color='k')
             
plt.scatter(hits[isNoises,0],hits[isNoises,1],marker='s',s=10,c='grey', alpha=0.8)
plt.scatter(hits[8,0],hits[8,1],marker='*',s=800,c='g', linewidth=0.5,edgecolor="k",alpha=0.8)
plt.scatter(hits[17,0],hits[17,1],marker='*',s=800,c='r', linewidth=0.5,edgecolor="k",alpha=0.8)

ax.grid(True,color='grey',linestyle="--",alpha=0.4)
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.tick_params(axis='y',labelsize=0,labelcolor="w")
ax.tick_params(axis='x',labelsize=0,labelcolor="w")
ax.text(-5,4.5,r"$d$",fontsize=18)

plt.savefig("../plots/Figure2.pdf")
plt.savefig("../plots/Figure2.png",dpi=400)
```
