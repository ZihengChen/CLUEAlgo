```python
from pylab import *
import pandas as pd
from sklearn import datasets
%matplotlib inline
```

# Figure 3

```python
fig, axes = plt.subplots(1,4,figsize=(16,4),facecolor='w')
fig.subplots_adjust(wspace=0.08)

fileDir = "../data/output/"

fileNames = ["aniso_1000_20_20_50_50",
             "sissa_1000_20_20_80_25",
             "moons_1000_20_20_150_15",
             "circles_1000_20_20_90_12"
            ]
footprints = [r"$a$",r"$b$",r"$c$",r"$d$"]
for i,fileName in enumerate(fileNames):
    df = pd.read_csv(fileDir+fileName+".csv")
    
    ax = axes[i]
    ax.set_aspect('equal')
    temp = df[df.clusterId==-1]
    ax.scatter(temp.x,temp.y, c='k', marker='.',s=1)
    temp = df[df.clusterId>-1]
    ax.scatter(temp.x,temp.y, c=temp.clusterId, marker='.', cmap='tab10',vmax=9,vmin=0,s=20)
    for index, row in temp.iterrows():
        row_nh = temp[temp.index==row.nh]
        if len(row_nh)==1:# not seed
            plotx = [row.x,row_nh.x]
            ploty = [row.y,row_nh.y]
            ax.plot(plotx,ploty,c="k",lw=0.5,alpha=1)
    temp = df[df.isSeed==1]
    ax.scatter(temp.x, temp.y, c='darkred',marker='*',s=200,alpha=1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.text(-240,-240,footprints[i],fontsize=18)

plt.savefig("../plots/Figure3.pdf")
plt.savefig("../plots/Figure3.png",dpi=400)
```

# Figure 4

```python
fig, axes = plt.subplots(1,4,figsize=(16,4),facecolor='w')
fig.subplots_adjust(wspace=0.08)

fileDir = "../data/output/"

fileNames = ["sissa_4000_20_10_80_80","sissa_4000_20_20_80_80","sissa_4000_20_60_80_80",]
footprints = [r"$a$",r"$b$",r"$c$",r"$d$"]
footprints2 = ["",r"$\delta_o=10$",r"$\delta_o=20$",r"$\delta_o=60$"]

ax = axes[0]
ax.set_aspect(40/9)
df = pd.read_csv(fileDir+fileNames[0]+".csv")
delta = df.delta.values.copy()
delta[df.delta==999] = 80

ax.scatter(df.rho,delta, marker='.',color='k', s=1)
ax.scatter(df[df.isSeed==1].rho, delta[df.isSeed==1], c='darkred',marker='*',s=200,alpha=1)
ax.axvline(80,color='C0')
ax.text(85,50,r"$\rho_c=80$",rotation='vertical',color='C0',fontsize=12)

ax.axhline(20,80/400,1,color='C0')
ax.text(300,22,r"$\delta_c=20$",color='C0',fontsize=12)

ax.axhline(10,0,80/400,color='C1',linestyle="--",linewidth=3)
ax.axhline(20,0,80/400,color='C1',linestyle="--",linewidth=3)
ax.axhline(60,0,80/400,color='C1',linestyle="--",linewidth=3)
ax.text(5,12,r"$\delta_o=10$",color='C1')
ax.text(5,22,r"$\delta_o=20$",color='C1')
ax.text(5,53,r"$\delta_o=60$",color='C1')

ax.set_xlim(0,400)
ax.set_ylim(0,90)
ax.set_xlabel(r"$\rho$",fontsize=14)
ax.set_ylabel(r"$\delta$",fontsize=14)
ax.tick_params(axis='y',labelsize=12)
ax.tick_params(axis='x',labelsize=12)
ax.text(20,81,footprints[0],fontsize=18)
ax.set_title("Decision Plot")


for i,fileName in enumerate(fileNames):
    df = pd.read_csv(fileDir+fileName+".csv")
    
    ax = axes[i+1]
    ax.set_aspect('equal')
    temp = df[df.clusterId==-1]
    ax.scatter(temp.x,temp.y, c='k', marker='.',s=1)
    temp = df[df.clusterId>-1]
    ax.scatter(temp.x,temp.y, c=temp.clusterId, marker='.', cmap='tab10',vmax=9,vmin=0,s=20)
    temp = df[df.isSeed==1]
    ax.scatter(temp.x, temp.y, c='darkred',marker='*',s=200,alpha=1)
    ax.set_xlim(-250,250)
    ax.set_ylim(-250,250)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.text(-240,200,footprints[i+1],fontsize=18)
    ax.text(100,-240,footprints2[i+1],fontsize=14,color='C1')

plt.savefig("../plots/Figure4.pdf")
plt.savefig("../plots/Figure4.png",dpi=400)
```
